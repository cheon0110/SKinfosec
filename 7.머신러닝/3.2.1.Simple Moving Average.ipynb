```python
import pandas as pd
```

```python
# W : 일요일들이 리턴됨 첫번째 일요일,두번째 일요일...
print(pd.date_range(start="2021-08", periods=8, freq="W"))

# M : 모든 월의 마지막날
print(pd.date_range(start="2021-08", periods=8, freq="M"))

# MS : 해당 월의 첫번째날
print(pd.date_range(start="2021-08", periods=8, freq="MS"))

# BMS : 해당 월의 첫번째 평일
print(pd.date_range(start="2021-08", periods=8, freq="BMS"))

# 2021년 7일 주기로 매주 월요일
# 7*'1d'= 7일
print(pd.date_range(start="2021-08-02", periods=8, freq=7*"1d"))
```

```python
# 196
import numpy as np

df = pd.DataFrame({
    "week":pd.date_range(start="2021-08-02", periods=8, freq=7*"1d"), # 매주 월요일
    "sales":[39, 44, 40, 45, 38, 43, 39, np.nan],
    "3MA":[0]*8
})

df
```

```python
# 3개의 행의 평균
df[["sales"]].rolling(3).mean().shift(1)

# 최근 3주의 평균을 계산 후 아래로 1줄 이동
# 예측 한 값 3MA에 저장
df["3MA"] = df[["sales"]].rolling(3).mean().shift(1)
df
```
