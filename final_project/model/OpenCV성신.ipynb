Haar Cascade -> 기계학습을 이용한 객체 탐지 알고리즘

CNN,RSTM등의 신경망이 아닌, 하르 특징(Haar Feature)계산을 통해 객체를 찾음

사전 학습된 분류를 로드 한 뒤, cv2통해 분류기 생성

# Haat cascade 분류기 불러오기

```python
# 가중치 파일 경로
cascade_filename = 'haarcascade_frontalface_alt.xml'

# 모델 불러오기
cascade = cv2.CascadeClassifier(cascade_filename)
```

# Test영상 사진 불러오기

```python
# 영상 파일
cam = cv2.VideoCapture('sample.mp4')

# 사진 파일
img = cv2.imread('sample.jpg',cv2.IMREAD_COLOR)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 영상 재생
def videoDetector(cam,cascade):
  while True:

    # 캡쳐이미지 불러오기
    ret, img = cam.read()
    # 영상 압축 
    img = cv2.resize(img,dsize=None,fx=0.75,fy=0.75)
    # 그레이 스케일 변환
    # 흑백 영상으로 변환
    # cvtColor 통해 영상 전처리를 진행
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)


    results = cascade.detectMultiScale(gray, # 입력 이미지
                                      scaleFactor=1.5, # 이미지 피라미드 스케일 Factor
                                      minNeighbors = 5, # 인접객체 최소 거리 픽셀
                                      minSize=(20,20)) # 탐지 객체 최소 크기

    print(results.shape)

    # 결괏값
    for x,y,w,h in results:
      cv2.rectangle(img, (x,y), (x+w,y+h),(255,0,0), 2)      

    cv2_imshow(img)   
              
videoDetector(cam,cascade)



# 사진 출력
def imgDetector(img,cascade):
  img = cv2.resize(img,dsize=None,fx=0.5,fy=0.5)
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

#얼굴 탐지 알고리즘 적용

```python
img = cv2.imread('images.jpg',cv2.IMREAD_COLOR)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# cascade 얼굴 탐지 알고리즘
# detectMultiScale : 객체 탐지 알고리즘 실행
# scaleFactor : 변화 정도율
results = cascade.detectMultiScale(gray, # 입력 이미지
                                   scaleFactor=1.5, # 이미지 피라미드 스케일 Factor
                                   minNeighbors = 5, # 인접객체 최소 거리 픽셀
                                   minSize=(20,20)) # 탐지 객체 최소 크기

print(results.shape)

# 결괏값
for x,y,w,h in results:
  cv2.rectangle(img, (x,y), (x+w,y+h),(255,0,0), 2)      

cv2_imshow(img)   
              
```
