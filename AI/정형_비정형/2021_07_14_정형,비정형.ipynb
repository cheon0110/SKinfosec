```python
# loc  : 줄 이름으로 줄 선택 ["서준"] : 서준 줄 선택
# iloc : 줄 순서로 줄 선택 0부터 시작 / [0] 번째 줄 선택 
# i -> index라고 생각하면 쉬움 index는 숫자~

df.loc["서준"]
df.iloc[0]

df.loc[["서준", "우현"]] # loc 옵션 [ 이상 : 이하 ] = 서준 이상 우현 미만
df.iloc[0:2]          # iloc 옵션 [ 이상 : 미만 ] = 0 이상 2 미만
```

```python
# p45
import pandas as pd

exam_data = {
    '이름' : ['서준', '우현', '인아'],
    '수학' : [90, 80, 70],
    '영어' : [98, 89 ,95],
    '음악' : [85, 95, 100],
    '체육' : [100, 90, 90]
}
df =pd.DataFrame(exam_data)
df_set = df.set_index('이름')
print(df_set)
print("==================================================")

df_set.iloc[0][3] = 80
print(df_set)
print("==================================================")
df_set.loc['서준']['체육'] = 90
print(df_set)

print("==================================================")
df_set.loc['서준','체육'] = 100
print(df_set)
print("==================================================")

df_set.loc['서준', ["음악", "체육"]] = 50
print(df_set)
print("==================================================")
```
