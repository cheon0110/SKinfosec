# 선형 회귀 분석(Gradient Decent  이용)
[ 선형 회귀 분석이란? ]
- 두 변수(종속, 독립) 사이의 함수적 관계를 기술하는 수학석 방정식
- 독립변수의 값이 주어질 때 종속변수의 값을 추정하거나 예측하는데 사용
- 서로 영향을 주고 받는 상관관계를 갖는 두 변수 사이의 관계를 분석
- 독립 변수 : 다른 변수에 영향을 주고 그 변수의 값을 예측 하려는 변수
- 종속 변수 : 독립 변수의 특정한 값에 따른 그의 값을 예측하고자 하는 변수

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
X = np.array([1,2,3], dtype="float32")
print("입력값 = ", X) # 입력값 X*w+b = 예측값(y^_통계, hypothesis_딥러닝)
y = np.array([2,2.5,3.5], dtype="float32")
print("입력값 = ",y) # 진짜 값
```

image.png

## w,b가 얼마일때 cost가 가장 최소가 될까?



image.png

```python
w = 2
b = 1
# 예측값 
hypothesis = X * w + b 
print("예측값 = ", hypothesis)
hypothesis - y # 예측값 - 진짜
```

image.png

## X, y 기울기 구하기

```python
# w의 기울기
w_gred = 1/3*np.dot((hypothesis - y),X)
w_gred

w = w - w_gred
print(" w의 기울기 = ", w)

# b의 기울기
b_gred = 1/3*(np.sum(hypothesis - y))
b_gred

b = b - b_gred
print(" b의 기울기 = ", b)
```

```python

# cost값이 줄어들어야 하는데 줄어들지 않음
# 오차 제곱의 합이 0에 가까운 값을 갖는 w와 b를 계산하지 못함
# =  기울기인 w_gred, b_gred가 너무 큰 값으로 계산이 되어 cost가 0이 되는 지점을 찾지 못함

for i in range(10):
  print("w = ", w) # w값 확인
  print("b = ", b)
  hypothesis = X*w+b
  print("hypothesis = ", hypothesis)
  print("y = ", y)
  cost = np.sum((hypothesis - y)**2) /3 # (예측 - 진짜)^의 합
  print("cost = ", cost)
  
  
  # w 기울기
  w_gred =1/3*np.dot((hypothesis - y), X)  
  print("w_gred = ", w_gred)

  # 업뎃
  w = w - w_gred

  # b 기울기
  b_gred = 1/3 * np.sum(hypothesis - y)
  print("b_gred = ", b_gred)
  b = b - b_gred
  
  print("="*100)
  print("업데이트 된 w = ", w)
  print("업데이트 된 b = ", b)
  print("="*100)
```

image.png

```python
w = 2
b = 1 

# 0.1을 곱해서 조금만 변하도록 함
learning_rate = 0.1
```

```python
# w_gred(5.5) * 0.1 = 0.55
# b_gred(2.3) * 0.1 = 0.233
# w = 2 - 0.55 = 1.45
# b = 1 - 0.233 = 0.766
for i in range(100):
  print("w = ", w) # w값 확인
  print("b = ", b)
  hypothesis = X*w+b
  print("hypothesis = ", hypothesis)
  print("y = ", y)
  cost = np.sum((hypothesis - y)**2) /3 # (예측 - 진짜)^의 합
  print("cost = ", cost)
  
  
  # w 기울기
  w_gred =(1/3*np.dot((hypothesis - y), X)) * learning_rate  
  print("w_gred = ", w_gred)
  w = w - w_gred # 업데이트

  # b 기울기
  b_gred = (1/3 * np.sum(hypothesis - y)) * learning_rate
  print("b_gred = ", b_gred)
  b = b - b_gred # 업데이트

  print("="*100)
  print("업데이트 된 w = ", w)
  print("업데이트 된 b = ", b)
  print("="*100)
  
```
